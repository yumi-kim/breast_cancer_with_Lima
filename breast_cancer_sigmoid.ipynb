```python
'''

2019. 5.27.
by Yumi Kim CBBL at GNU
Logistic Regression Classifier(Using Sigmoid) (0 or 1)
running rate = 0.01
total epoch = 10001
up to these parameters, the accuracy can be changed.
'''

# import libraries

import tensorflow as tf
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

# Read from csv file
# the data after optimization
raw_dataset = pd.read_csv('Breast_Cancer_Coimbra_Data_Set.csv')

# Set for dataSet
raw_array = raw_dataset.values
#random.shuffle(raw_array)

# Data Scaling
scaler = MinMaxScaler(feature_range=(0, 1)) 
raw_array = scaler.fit_transform(raw_array)

col_length = raw_array.shape[1]-1

x_data = raw_array[:,0:-1]
y_data = raw_array[:,[-1]]

# Data split for Train and Test Data 7 : 3
X_train, X_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.3, random_state=1)


# for reproducibility
tf.set_random_seed(777)  

# placeholders for a tensor that will be always fed.

X = tf.placeholder(tf.float32, shape=[None, col_length])

Y = tf.placeholder(tf.float32, shape=[None, 1])


W = tf.Variable(tf.random_normal([col_length, 1]), name='weight')

b = tf.Variable(tf.random_normal([1]), name='bias')

# Hypothesis using sigmoid: tf.div(1., 1. + tf.exp(tf.matmul(X, W)))

hypothesis = tf.sigmoid(tf.matmul(X, W) + b)

# cost/loss function

cost = -tf.reduce_mean(Y * tf.log(hypothesis) + (1 - Y) *

                       tf.log(1 - hypothesis))


train = tf.train.GradientDescentOptimizer(learning_rate=0.01).minimize(cost)


''' 
 Accuracy computation

 True if hypothesis>0.5 else False

'''

predicted = tf.cast(hypothesis > 0.5, dtype=tf.float32)

accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, Y), dtype=tf.float32))



'''
 Launch graph

'''

with tf.Session() as sess:

    # Initialize TensorFlow variables

    sess.run(tf.global_variables_initializer())


    for step in range(20001):

        cost_val, _ = sess.run([cost, train], feed_dict={X: X_train, Y: y_train})

        if step % 200 == 0:

            print(step, cost_val)



    # Accuracy report for Test Data

    h, c, a = sess.run([hypothesis, predicted, accuracy],

                       feed_dict={X: X_test, Y: y_test})

    print("\nW: ", sess.run(W), " bias: ", sess.run(b))
    print("\nHypothesis: ", h, "\nCorrect (Y): ", c, "\nAccuracy: ", a)
```
